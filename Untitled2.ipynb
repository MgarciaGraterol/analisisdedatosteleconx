```python
# 1️⃣ Importar librerías necesarias
import pandas as pd

# 2️⃣ Cargar el archivo JSON
# pd.read_json es ideal para archivos con estructura JSON compatible
df = pd.read_json("TelecomX_Data.json")

# 3️⃣ Verificar las primeras filas para entender la estructura
print("Vista inicial de los datos:")
df.head()
```

```python
# 2️⃣ Aplanar columnas anidadas
# Expandimos las columnas de tipo diccionario y las unimos al dataframe principal
customer_df = pd.json_normalize(df["customer"])
phone_df = pd.json_normalize(df["phone"])
internet_df = pd.json_normalize(df["internet"])
account_df = pd.json_normalize(df["account"])

# Unimos todo en un solo dataframe
df_flat = pd.concat([df.drop(columns=["customer", "phone", "internet", "account"]),
                     customer_df, phone_df, internet_df, account_df], axis=1)

# 3️⃣ Verificamos valores ausentes
print("Valores nulos por columna:")
print(df_flat.isnull().sum())

# 4️⃣ Verificamos duplicados
duplicados = df_flat.duplicated().sum()
print(f"\nFilas duplicadas: {duplicados}")

# 5️⃣ Detectar errores de formato
# Ejemplo: columnas numéricas que deberían serlo
print("\nTipos de datos:")
print(df_flat.dtypes)

# 6️⃣ Verificar inconsistencias en categorías (solo para columnas de tipo object)
for col in df_flat.select_dtypes(include="object").columns:
    print(f"\nColumna: {col}")
    print(df_flat[col].value_counts(dropna=False))
```

```python

# 2️⃣ Eliminar duplicados (si existen)
df_flat.drop_duplicates(inplace=True)

# 3️⃣ Manejar valores nulos
# - Para columnas categóricas: reemplazamos NaN con "Unknown"
# - Para columnas numéricas: reemplazamos con la mediana (menos sensible a outliers)
for col in df_flat.columns:
    if df_flat[col].dtype == "object":
        df_flat[col].fillna("Unknown", inplace=True)
    else:
        df_flat[col].fillna(df_flat[col].median(), inplace=True)

# 4️⃣ Convertir tipos de datos correctos
# Ejemplo: SeniorCitizen y tenure deberían ser enteros
if "SeniorCitizen" in df_flat.columns:
    df_flat["SeniorCitizen"] = df_flat["SeniorCitizen"].astype(int)
if "tenure" in df_flat.columns:
    df_flat["tenure"] = df_flat["tenure"].astype(int)

# 5️⃣ Estandarizar valores categóricos
# Ejemplo: 'Yes', 'No', 'no', 'YES' → 'Yes' o 'No'
for col in df_flat.select_dtypes(include="object").columns:
    df_flat[col] = df_flat[col].str.strip().str.title()

# 6️⃣ Corrección de categorías inconsistentes en servicios
# Si hay valores como 'No internet service' o 'No phone service', los pasamos a 'No'
cols_servicios = [
    "OnlineSecurity", "OnlineBackup", "DeviceProtection",
    "TechSupport", "StreamingTV", "StreamingMovies",
    "MultipleLines", "InternetService", "PhoneService"
]
for col in cols_servicios:
    if col in df_flat.columns:
        df_flat[col] = df_flat[col].replace(
            {"No Internet Service": "No", "No Phone Service": "No"}
        )

# 7️⃣ Confirmar limpieza
print("Valores nulos después de limpieza:")
print(df_flat.isnull().sum())

print("\nMuestra de datos limpios:")
df_flat.head()
```

```python
# 1️⃣ Verificar que la columna MonthlyCharges exista
if "Charges.Monthly" in df_flat.columns:

    # Aseguramos que MonthlyCharges sea numérica
    df_flat["Charges.Monthly"] = pd.to_numeric(df_flat["Charges.Monthly"], errors="coerce")

    # Calcular valor diario (aproximando a 30 días por mes)
    df_flat["Cuentas_Diarias"] = df_flat["Charges.Monthly"] / 30

    # Redondear a 2 decimales
    df_flat["Cuentas_Diarias"] = df_flat["Cuentas_Diarias"].round(2)

    print("✅ Columna 'Cuentas_Diarias' creada correctamente.")
    print(df_flat[["Charges.Monthly", "Cuentas_Diarias"]].head())

else:
    print("⚠️ La columna 'Charges.Monthly' no se encontró en el dataset.")
```

```python
# 1️⃣ Análisis descriptivo general de todas las columnas numéricas
print("📊 Resumen estadístico general:\n")
print(df_flat.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1️⃣ Gráfico de torta (proporción)
churn_counts = df_flat["Churn"].value_counts()
plt.figure(figsize=(6,6))
plt.pie(churn_counts, labels=churn_counts.index, autopct='%1.1f%%', colors=["#4CAF50", "#F44336"], startangle=90)
plt.title("Proporción de clientes por estado de Churn", fontsize=14)
plt.show()

# 2️⃣ Gráfico de barras (conteo)
plt.figure(figsize=(6,4))
sns.countplot(data=df_flat, x="Churn", palette={"No": "#4CAF50", "Yes": "#F44336"})
plt.title("Distribución de clientes por estado de Churn", fontsize=14)
plt.xlabel("Churn")
plt.ylabel("Cantidad de clientes")
plt.show()

```

```python
# Aseguramos que las columnas estén en formato numérico
if "TotalCharges" in df_flat.columns:
    df_flat["TotalCharges"] = pd.to_numeric(df_flat["TotalCharges"], errors="coerce")
if "tenure" in df_flat.columns:
    df_flat["tenure"] = pd.to_numeric(df_flat["tenure"], errors="coerce")

# Lista de variables numéricas a analizar
variables_numericas = ["TotalCharges", "tenure", "MonthlyCharges", "Cuentas_Diarias"]

# Configuración de estilo
sns.set_style("whitegrid")

# 1️⃣ Boxplots para comparar distribución entre clientes que cancelaron y no cancelaron
for var in variables_numericas:
    if var in df_flat.columns:
        plt.figure(figsize=(6,4))
        sns.boxplot(data=df_flat, x="Churn", y=var, palette={"No": "#4CAF50", "Yes": "#F44336"})
        plt.title(f"Distribución de {var} según Churn", fontsize=14)
        plt.xlabel("Churn")
        plt.ylabel(var)
        plt.show()

# 2️⃣ Violinplots para ver forma de la distribución
for var in variables_numericas:
    if var in df_flat.columns:
        plt.figure(figsize=(6,4))
        sns.violinplot(data=df_flat, x="Churn", y=var, palette={"No": "#4CAF50", "Yes": "#F44336"})
        plt.title(f"Distribución detallada de {var} según Churn", fontsize=14)
        plt.xlabel("Churn")
        plt.ylabel(var)
        plt.show()
```
